## Cleaning Functions Twitter
Functions that will be used to clean and anonymize our data. We want to get rid of @mentions, #hashtags, links in tweet text and extract them into separate columns and anonymize users. We saved usernames so we can try and analyse if there are some propaganda bots. Same username will always have same hash.

```python
def anonymize(data):
    """Converts input into sha256 hash"""
    data = bytes(data, 'utf-8')
    hash = sha256()
    hash.update(data)
    return hash.hexdigest()


def get_links(text):
    """Extracts links from tweet text"""
    words = text.split()
    
    links = []
    for word in words:
        if word.startswith('https'):
            links.append(word)
    return links
    
    
def get_hashtags(text):
    """Extracts hashtags from tweet text"""
    words = text.split()
    
    hashtags = []
    for word in words:
        if word.startswith('#'):
            hashtags.append(word.lstrip('#')) # removes '#' from beginnig so we are left with only hashtag word  
    return hashtags


def remove_mentions_links_tags(text):
    """Removes mentiones, links and hashtags from tweet text"""
    
    if text is None: # in order for select method to work we must escape NoneType case
        return 0
    
    words = text.split(' ')
    cleaned = []
    for word in words:
        if '@' in word or word.startswith('https') or word.startswith('#') or word == 'RT':
            continue
        else:
            cleaned.append(word)
    return ' '.join(cleaned)


def trump_or_biden(text):
    if isinstance(text, str):
        r = re.compile(r'[\s{}]+'.format(re.escape(string.punctuation)))
        words = r.split(text)
        words = text.split()

        for word in words:
            word = word.lower()
            if word == 'trump' or word == "trump's":
                return 'Trump'
            elif word == 'biden' or word == "biden's":
                return 'Biden'
            else:
                continue


# Sentiment Functions
def apply_blob(text):
    if isinstance(text, str):
        return TextBlob(text).sentiment[0]


def apply_vader(text):
    if isinstance(text, str):
        return SentimentIntensityAnalyzer().polarity_scores(text)['compound']
```

#### Making cleaning functions as User Defined Functions so they work on spark DataFrames


```python
remove_udf = udf(remove_mentions_links_tags, StringType())
anonymize_udf = udf(anonymize, StringType())
blob_udf = udf(apply_blob, StringType())
get_links_udf = udf(get_links, StringType())
get_hashtags_udf = udf(get_hashtags, StringType())
vader_udf = udf(apply_vader, StringType())
who_udf = udf(trump_or_biden, StringType())
```

#### Applying our cleaning and preparation functions

```python
df1 = df.withColumn('links', get_links_udf(df['tweet_text'])) # extracting links
df2 = df1.withColumn('oriented', who_udf(df1['tweet_text'])) # extracting who is mentioned first
df3 = df2.withColumn('tags', get_hashtags_udf(df2['tweet_text'])) # hashtags
df4 = df3.withColumn('tweet_text', remove_udf(df3['tweet_text'])) # cleaning the text
df5 = df4.withColumn('user_name', anonymize_udf(df4['user_name'])) # anonymizing users
df6 = df5.withColumn('text_blob', blob_udf(df4['tweet_text'])) # applying textblob to tweet text
df6 = df6.withColumn("date_only", F.to_date(F.col("tweet_created_at"))) # overwriting tweet_created_at so it shows only the date without time -> easier for later manipulation
```

```python
# here we can adjust text_blob results

df_blob = df6.withColumn(
    'blob_rated', # chained when
    F.when((F.col("text_blob") < 0.0), 'Negative')\
    .when((F.col("text_blob") > 0.0) , 'Positive')\
    .otherwise('Neutral')
)

# Vader is too slow, we skip it for now
# final = df7.withColumn(
#     'vader_rated', # chained when
#     F.when((F.col("text_vader") < -0.5), 'Negative')\
#     .when((F.col("text_vader") > 0.5) , 'Positive')\
#     .otherwise('Neutral')
# )
```

## VISUALISATION TWITTER

This is our function for plotting barcharts between two different time intervals or between two other categories. It is reusable and makes exploring much easier.

```python
# function to plot difference between two days - two categories
def plot_difference(title, df1, df2, column, file_to_save, title_one, title_two):
    order = ['Positive', 'Neutral', 'Negative']
    palette = ['#00ff00', '#0000ff','#ff0000'] # green-positive, blue-neutral, red-negative

    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15,5))
    fig.suptitle(title, fontsize=16)
    sns.set_style("dark")
    sns.despine(left=True, bottom=True)
    sns.countplot(ax=ax1,
                  x=column,
                  data=df1,
                  palette=palette, 
                  order=order) 

    sns.countplot(ax=ax2,
                  x=column,
                  data=df2,
                  palette=palette,
                  order=order)

    ax1.set_title(title_one)
    ax1.set_ylabel('')
    ax1.set_xlabel('')

    ax2.set_title(title_two)
    ax2.set_ylabel('')
    ax2.set_xlabel('')


    fig.show()
    plt.savefig('./data/graphs/' + file_to_save)
```

### Scatter-text Visualisation
Here we will implement one of the most fascinating and interesting libraries we came across so far. Its scatter-text, it produces really amazing, interactive and very insightful scatter plots that allow us to better understand word usage of two sides.

```python
df = df_blob.filter("followers > 10000 and (oriented == 'Trump' or oriented == 'Biden')").toPandas()
```

```python
nlp = st.whitespace_nlp_with_sentences
corpus = st.CorpusFromPandas(df,
                              category_col='oriented',
                              text_col='tweet_text',
                              nlp=nlp).build()


file_name = './data/graphs/scatter-txt.html'

html = st.produce_scattertext_explorer(corpus,
                                       category='Biden',
                                       category_name='Biden mentioned',
                                       not_category_name='Trump mentioned',
                                       width_in_pixels=1000,
                                       metadata=df['oriented']
                                       )

file = open(file_name, 'wb')
file.write(html.encode('utf-8'))
file.close()

# NOTEBOOK WILL FREEZE BUT JUST WAIT SOMETIME UNTIL IT LOADS THE PLOT
display(HTML("<style>.container { width:98% !important; }</style>"))
IFrame(src=file_name, width = 1200, height=700)
```

```python
df = reddit_df.toPandas()
```

```python
corpus = ' '.join(df['comment'])
splited = re.findall(r"[\w']+", corpus) # spliting on new lines, whitespaces and everything else that might be seen useful to split on

print(splited[:100])
```

```python
# we build frequency dictionary out of corpus, so it means the bigger the word the more times it was used in comments
freq = {}
for word in splited:
    word = word.lower()
    if word not in freq.keys():
        freq[word] = 1
    else:
        freq[word] += 1


# adjusting stopwords
STOPWORDS.add('https')
STOPWORDS.add('The')
STOPWORDS.add('s')
STOPWORDS.add('t')
# removing stopwords from freq. dict
for key in freq.keys():
    if key in STOPWORDS:
        freq[key]= 0
        
print('Trump:', freq['trump'])
print('Biden:', freq['biden'])
print('Republicans:', freq['republicans'])
print('Democrats:', freq['democrats'])
```

```python
# creating the wordcloud
wordcloud = WordCloud(
    width = 2000,
    height = 1000,
    background_color = 'white').generate_from_frequencies(freq)


fig = plt.figure(
    figsize = (20, 10),
    facecolor = 'k',
    edgecolor = 'k')

plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0) # no borders when saving to file
plt.savefig('./data/graphs/reddit-wordcloud.png')
plt.show()
```
